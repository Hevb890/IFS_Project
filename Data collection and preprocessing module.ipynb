```python
import requests
from bs4 import BeautifulSoup

def scrape_topic_content(base_url, start_page, end_page):
    topic_content = []

    try:
        for page_num in range(start_page, end_page + 1):
            page_url = f"{base_url}/index{page_num}.html"
            response = requests.get(page_url, headers={'User-Agent': 'Mozilla/5.0'})
            response.raise_for_status()

            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract topic links from the current page
            topic_links = soup.find_all('a', class_='topic-view_link')
            
            # Scrape content for each topic link
            for topic_link in topic_links:
                topic_url = community_url + topic_link['href']
                response_topic = requests.get(topic_url, headers={'User-Agent': 'Mozilla/5.0'})
                response_topic.raise_for_status()

                soup_topic = BeautifulSoup(response_topic.content, 'html.parser')
                content_element = soup_topic.find('div', class_='post__content js-content--original qa-topic-post-content post__content--new-editor')

                content_text = content_element.get_text(strip=True) if content_element else "No content found"

                # Append the content to the list
                topic_content.append({'url': topic_url, 'content': content_text})

    except Exception as e:
        print(f"Error during web scraping: {e}")

    return topic_content

# Example usage
base_community_url = "https://community.ifs.com/supply-chain-251"
community_url = "https://community.ifs.com"
start_page_number = 1
end_page_number = 4

supply_chain_topic_content = scrape_topic_content(base_community_url, start_page_number, end_page_number)

# Print the scraped topic content
for index, topic_content in enumerate(supply_chain_topic_content, start=1):
    print(f"Topic {index} - URL: {topic_content['url']}")
    print(f"Content:\n{topic_content['content']}\n{'-'*50}\n")
```
